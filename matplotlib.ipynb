```python
# red, 별모양, dash선으로 sin 그리기
# green, 동그라미, 실선으로 cos 그리기
theta = np.linspace(0, 2*np.pi, 20)
#print(theta)
plt.plot(theta, np.sin(theta), 'r*--', markersize=10)
#plt.figure() # 새창 열기 =
plt.plot(theta, np.cos(theta), 'go-', label="cos")
plt.xlabel(r"$\theta$") # r은 raw
plt.ylabel("value")
plt.title("sin and cos", color="b")
plt.legend() # 레이블 출력
plt.grid() # 격자
#plt.xlim([0, np.pi]) # 구간
#plt.ylim(-1.5, 1.5)
plt.axis([0, np.pi, -1.5, 1.5]) # 위에 두 줄이랑 같음
```

```python
x=range(5)
y1 = [i for i in range(1, 6)]
y2 = [i + 100 for i in range(1,6)]

plt.figure()
plt.plot(x, y1)
plt.plot(x, y2)


plt.figure()
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
p1 = ax1.plot(x, y1, 'b')
p2 = ax2.plot(x, y2, 'r')

ax1.set_ylim([-10, 10])

ax1.set_xlabel("x")
ax1.set_ylabel("y1")
ax2.set_ylabel("y2")
ax1.set_title("y1 and y2")
ax1.grid()

plt.legend(p1+p2, ["y1", "y2"], loc="best") # location
```

# 2개 변수 함수, 등고선으로 보기, 3D로 보기

```python
a = np.linspace(-1, 1, 100)
b = np.linspace(-1, 1, 100)
# a^2 + b^2
A, B = np.meshgrid(a,b)

L = A**2 + B**2
plt.contour(a, b, L, 20) # 등고선 , 20-> 등고선 개수
plt.xlabel("a"); plt.ylabel("b"); plt.grid()
```

```python
plt.figure(figsize =[10, 9]) # 가로, 세로 크기
ax = plt.axes(projection="3d") # 3d 그리기
ax.plot_surface(A, B, L)

plt.figure(figsize=[10, 9])
ax = plt.axes(projection="3d")
ax.plot_wireframe(A, B, L)
```

```python
a = np.linspace(-1, 1, 10)
b = np.linspace(-1, 1, 10)
A, B = np.meshgrid(a,b)
L = A**2 + B**2

plt.figure(figsize=[10, 9])
ax = plt.axes(projection="3d")
plt.plot(A.reshape(-1, 1).squeeze(), B.reshape(-1, 1).squeeze(),
         L.reshape(-1, 1).squeeze(), "ro") # 1차원으로

plt.figure(figsize=[10, 9])
ax = plt.axes(projection="3d")
#ax.scatter(A, B, L, s=100, c='g', marker="*", edgecolor="r") # 2차원 가능 , 사이즈, 컬러
ax.scatter(A, B, L, s=100, c='g', marker="*", alpha=0.3)  # 투명도
```

```python
plt.figure(figsize=[10, 9])
ax = plt.axes(projection="3d")
ax.plot_surface(A, B, L, alpha=0.3, cmap="viridis")
ax.scatter(A, B, L, s=100, c="r", marker="*")
```

# plotly


```python
import plotly.graph_objects as go

a=np.linspace(-2, 2, 100)
b=np.linspace(-2, 2, 100)
A, B = np.meshgrid(a, b)
L = np.cos(A**2.4) + np.sin(B**2.4)

fig = go.Figure(data=[go.Surface(x=a, y=b, z=L, colorscale="viridis", opacity=0.5)])
fig.update_traces(contours_z=dict(show=True, usecolormap=True, highlightcolor="limegreen",
                                  project_z=True))
fig.update_layout(title="cos(A^2.4)+sin(B^2.4)", width=700, height=600)
```

```python
fig = go.Figure(data=[go.Surface(x=a, y=b, z=L, opacity=0.5)])
fig.update_traces(contours_z=dict(show=True, usecolormap=True, highlightcolor="limegreen",
                                  project_z=True))
fig.update_layout(title="cos(A^2.4)+sin(B^2.4)", width=700, height=600,
                  scene=dict(zaxis=dict(nticks=50, range=[-3, 3])))
```

```python
import matplotlib.pyplot as plt
from torchvision import datasets
test_DS = datasets.MNIST(root='data/', train=False, download=True)
```

```python
print(test_DS.data.shape)
print(test_DS.data[0])
```

```python
plt.imshow(test_DS.data[0]) # image show
```

```python
plt.imshow(test_DS.data[8], cmap="gray")
plt.colorbar()
```

```python
for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.imshow(test_DS.data[i], cmap="gray")
```

```python
plt.figure(figsize=[10, 9])
for i in range(6):
    plt.subplot(2, 3, i+1, yticks=[])
    plt.xticks([])
    plt.imshow(test_DS.data[i], cmap="gray")
```

```python
plt.figure(figsize=[7, 4])
for i in range(6):
    plt.subplot(2, 3, i+1, yticks=[] ,xticks=[])
    plt.imshow(test_DS.data[i], cmap="gray")
    plt.title(test_DS.targets[i].item(), color="g")
```

```python
im = test_DS.data[0]
plt.figure(figsize=(20, 15))
plt.imshow(im, cmap="gray")
plt.colorbar()

for i in range(im.shape[0]):
    for j in range(im.shape[1]):
        plt.text(j, i, im[i,j].item(),
                 horizontalalignment="center",
                 color="k" if im[i,j] > torch.max(im)/1.5 else "w")

plt.xticks([])
plt.yticks([])
```
