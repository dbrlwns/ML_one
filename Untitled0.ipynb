```python
import numpy as np
import matplotlib.pyplot as plt

x = [150, 160, 170, 175, 185] # 키
y = [55, 70, 64, 80, 75] # 몸무게

plt.plot(x, y, 'o')
```

```python
a = 0.45
b = -35
x_plot = np.linspace(145, 190, 100)
y_plot = a * x_plot + b

plt.plot(x, y, 'o')
plt.plot(x_plot, y_plot, 'r')
```

```python
import numpy as np
a = 0.5 + np.linspace(-0.2, 0.2, 100)
b = -30 + np.linspace(-20, 20, 100)
A, B = np.meshgrid(a, b) # 100x100 2개

L = np.zeros_like(A) # 100x100

for xi, yi in zip(x, y):
    L += (yi-(A*xi+B)) ** 2

plt.figure()
plt.contour(a,b,L,30); plt.xlabel("a"); plt.ylabel("b"); plt.grid()

plt.figure(figsize=[10,9])
ax = plt.axes(projection="3d")
ax.plot_surface(A, B, L); plt.xlabel("a"); plt.ylabel("b")
ax.set_zlim([0,5000])

print(f"optimal a : {A[L == np.min(L)]}")
print(f"optimal b : {B[L == np.min(L)]}")
```

```python
a_opt = A[L==np.min(L)]
b_opt = B[L==np.min(L)]

plt.figure()
y_plot = a_opt * x_plot + b_opt
plt.plot(x, y, "o")
plt.plot(x_plot, y_plot, 'r')
```

```python
import torch
import matplotlib.pyplot as plt
x_plot = np.linspace(145, 190, 100)
a = torch.tensor(0.45, requires_grad=True)
b = torch.tensor(-35.0, requires_grad=True)

L=0
for xi, yi in zip(x,y):
    L += (yi - (a*xi + b))**2

L.backward()
print(a.grad, b.grad)

LR = 1e-6
a = a.detach() - LR * a.grad
b = b.detach() - LR * b.grad

print(a.item(), b)

plt.figure()
y_plot = a * x_plot + b
plt.plot(x, y, "o")
plt.plot(x_plot, y_plot, "r")
```

```python
a.requires_grad=True
b.requires_grad = True

L=0
for xi, yi in zip(x,y):
    L += (yi - (a*xi + b))**2

L.backward()
print(a.grad, b.grad)

LR = 1e-6
a = a.detach() - LR * a.grad
b = b.detach() - LR * b.grad

print(a.item(), b)

plt.figure()
y_plot = a * x_plot + b
plt.plot(x, y, "o")
plt.plot(x_plot, y_plot, "r")
```

```python
EPOCH = 10
LR = 1e-6
plt.plot(x, y, "o")

for _ in range(EPOCH):
    a.requires_grad=True
    b.requires_grad=True

    L=0
    for xi, yi in zip(x,y):
        L += (yi - (a*xi + b))**2

    L.backward()

    a = a.detach() - LR * a.grad
    b = b.detach() - LR * b.grad

    y_plot = a * x_plot + b
    plt.plot(x_plot, y_plot, "r")
```
